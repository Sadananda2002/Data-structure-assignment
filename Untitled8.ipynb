## 1. Discuss string slicing and provide examples.

     *String Slicing in Python:
      String slicing is a technique in Python to extract a part of a string (substring) by specifying a range of indices. The general syntax is string[start:stop:step],
      where: - *start*: The starting index of the slice (inclusive).
              - *stop*: The ending index (exclusive).
              - *step*: The increment (optional). By default, it's 1.

      Examples:
      python
      text = "Hello, World!"
      print(text[0:5])   # Output: 'Hello'
      print(text[:5])    # Output: 'Hello' (start index is optional)
      print(text[7:])    # Output: 'World!' (stop index is optional)
      print(text[::2])   # Output: 'Hlo ol!' (step of 2)
      print(text[::-1])  # Output: '!dlroW ,olleH' (reversed string)

      Python slicing can be done in two ways:

      Using a slice() method.
      Using the array slicing  [:: ] method.

## 2. Explain the key features of lists in Python.

     *Key Features of Lists in Python:*

      - *Mutable*: Lists can be modified (elements can be changed, added, or removed).
      - *Ordered*: Elements maintain their order of insertion.
      - *Heterogeneous*: Lists can contain elements of different types (integers, strings, etc.).
      - *Dynamic*: Lists can grow or shrink dynamically in size.
      - *Indexable*: Elements in a list can be accessed via indices.

      Example:
      python
      my_list = [1, "apple", 3.5, True]
   

## 3. Describe how to access, modify, and delete elements in a list with examples.

      Accessing, Modifying, and Deleting Elements in a List:*

      - *Accessing*: Use index positions.
      python
      my_list = [1, 2, 3, 4]
      print(my_list[0])  # Output: 1
      print(my_list[-1]) # Output: 4 (last element)
     
      - *Modifying*: Assign a new value to an index.
      python
      my_list[1] = 10
      print(my_list)  # Output: [1, 10, 3, 4]
     
      - *Deleting*: Use del, remove(), or pop().
      python
      del my_list[1]
      print(my_list)  # Output: [1, 3, 4]

      my_list.remove(3)
      print(my_list)  # Output: [1, 4]

      popped_item = my_list.pop()  # Removes the last element by default
      print(my_list)  # Output: [1]
     

## 4. Compare and contrast tuples and lists with examples.

      *Comparison Between Tuples and Lists:*
      - *Mutability*:

      - *List*: Mutable, meaning elements can be added, modified, or removed.
      - *Tuple*: Immutable, meaning once defined, the elements cannot be changed.

      - *Syntax*:

      - *List*: Defined using square brackets [].
      - *Tuple*: Defined using parentheses ().

      - *Usage*:

      - *List*: Used when you need a collection of items that may need modification.
      - *Tuple*: Used when you need a collection that shouldn’t change (e.g., fixed data like coordinates).

      Example:
      python
      my_list = [1, 2, 3]
      my_tuple = (1, 2, 3)
      # my_tuple[0] = 4  # This will raise an error because tuples are immutable.
   

## 5. Describe the key features of sets and provide examples.

      *Key Features of Sets:*

      - *Unordered*: Elements are not stored in any particular order.
      - *No Duplicates*: A set automatically removes duplicate elements.
      - *Mutable*: Elements can be added or removed.
      - *Heterogeneous*: Can contain elements of different data types.

      Example:
      python
      my_set = {1, 2, 3, 3, 4}
      print(my_set)  # Output: {1, 2, 3, 4} (no duplicates)

      *Use Cases of Tuples and Sets:*

      - *Tuples*:

      - Used when the data should not change, like geographical coordinates or constants.
      - Can be used as keys in dictionaries since they are immutable.

      - *Sets*:

      - Useful for membership testing (checking if an element exists) due to the high efficiency of set operations.
      - Ideal for removing duplicates from a list.
      - Supports mathematical operations like union, intersection, and difference.

 ## 7. Discribe how to add, modify, and delete items in a dictionary with examples.

      *Adding, Modifying, and Deleting Items in a Dictionary:*

      - *Adding/Modifying Items*: Dictionaries are key-value pairs, and new key-value pairs can be added, or existing keys can be updated.
      python
      my_dict = {"name": "Alice", "age": 25}
      my_dict["age"] = 26  # Modifies age
      my_dict["city"] = "New York"  # Adds new key-value pair
      print(my_dict)  # Output: {'name': 'Alice', 'age': 26, 'city': 'New York'}
     
      - *Deleting Items*: You can use del, pop(), or clear().
      python
      del my_dict["age"]  # Deletes the key 'age'
      print(my_dict)  # Output: {'name': 'Alice', 'city': 'New York'}

      city = my_dict.pop("city")  # Removes 'city' and returns its value
      print(my_dict)  # Output: {'name': 'Alice'}
      print(city)  # Output: 'New York'

      my_dict.clear()  # Clears all items from the dictionary
      print(my_dict)  # Output: {}
